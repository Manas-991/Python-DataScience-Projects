```python
from sklearn.model_selection import train_test_split,GridSearchCV
```

```python
X = df.drop('Churn_Yes',axis=1)
y = df['Churn_Yes']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
  X, y, test_size=0.1, random_state=101)
```

```python
from sklearn.ensemble import AdaBoostClassifier,GradientBoostingClassifier
```

```python
from sklearn.linear_model import LogisticRegressionCV
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
decision_tree_model = DecisionTreeClassifier(class_weight='balanced')
```

```python
decision_tree_model.fit(X_train,y_train)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay,classification_report
```

```python
ConfusionMatrixDisplay.from_estimator(decision_tree_model,X_test,y_test)
```

```python
y_pred_dtc = decision_tree_model.predict(X_test)
```

```python
print(classification_report(y_test,y_pred_dtc))
```

```python
log_model = LogisticRegressionCV(class_weight='Balanced')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
Scaled_X_train = scaler.fit_transform(X_train)
```

```python
Scaled_X_test = scaler.transform(X_test)
```

```python
log_model.fit(Scaled_X_train,y_train)
```

```python
ConfusionMatrixDisplay.from_estimator(log_model,Scaled_X_test,y_test)
```

```python
y_pred_lm = log_model.predict(Scaled_X_test)
```

```python
print(classification_report(y_test,y_pred_lm))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
help(RandomForestClassifier)
```

```python
param_grid_rfc = {'n_estimators':[100, 110, 120, 130, 140, 150, 160]
                  ,'max_features':["auto", "sqrt", "log2"],
                 'bootstrap':[True,False],
                 'class_weight':['none','balanced']
                 ,'warm_start':[True,False]}
```

```python
#np.linspace(100,200,11).tolist()
```

```python
grid_rfc = GridSearchCV(rfc,param_grid=param_grid_rfc)
```

```python
grid_rfc.fit(X_train,y_train)
```

```python
grid_rfc.best_params_
```

```python
y_pred_rfc = grid_rfc.predict(X_test)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred_rfc)
```

```python
print(classification_report(y_test,y_pred_rfc))
```
